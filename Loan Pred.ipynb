### Replacing Missing Values of Loan Aount and Credit History with the Mean and Median Respectively

```python
#Replacing Missing Values of Loan AMount with the Mean
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df['LoanAmount'].isnull().sum()
```

```python
df['Credit_History'].isnull().sum()
```

```python
#Replacing Missing Values of Credit History with the Median
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
```

```python
df['Credit_History'].isnull().sum()
```

```python
#Checking out for Entire Missing Values
df.isnull().sum()
```
